```python
master_data.loc[:, 'main_category'] = master_data['category_list'].apply(lambda x: x.split("|")[0])
master_data.head()
```

```python
master_data = master_data.drop('category_list', axis=1)
master_data.head()
```

```python
map_data = map_data[~pd.isnull(map_data['category_list'])]
map_data.isnull().sum()
```

```python
map_data['category_list'] = map_data['category_list'].str.lower()
master_data['main_category'] = master_data['main_category'].str.lower()
```

```python
master_data[~master_data['main_category'].isin(map_data['category_list'])]
```

```python
map_data[~map_data['category_list'].isin(master_data['main_category'])]
```

```python
map_data['category_list'] = map_data['category_list'].apply(lambda x: x.replace('0', 'na'))
print(map_data['category_list'])
```

```python
master_data = pd.merge(master_data, map_data, how='inner', left_on='main_category', right_on='category_list')
master_data.head()
```

```python
master_data = master_data.drop('category_list', axis=1)
master_data.head()
```

```python
master_data.info()
```

```python
value_vars = master_data.columns[10:19]

id_vars = np.setdiff1d(master_data.columns, value_vars)

print(value_vars, "\n")
print(id_vars)
```
